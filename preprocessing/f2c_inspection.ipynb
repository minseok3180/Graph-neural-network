```python
import csv
import math
from collections import defaultdict

def softmax(x):
    max_x = max(x)
    e_x = [math.exp(v - max_x) for v in x]
    total = sum(e_x)
    return [v / total for v in e_x]

data = defaultdict(lambda: {'compounds': [], 'weights': []})

with open(csv_path, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)
    for row in reader:
        weight = float(row['edge_weight'])
        # if weight == 0.0:
        #     continue
        food_id = int(row['food_id'])
        compound_id = int(row['drug_id'])
        data[food_id]['compounds'].append(compound_id)
        data[food_id]['weights'].append(weight)

# softmax   0.0  
threshold = 1e-8  #   
for food_id in list(data.keys()):
    weights = data[food_id]['weights']
    compounds = data[food_id]['compounds']
    softmax_weights = softmax(weights)

    filtered_compounds = []
    filtered_weights = []

    for c, w in zip(compounds, softmax_weights):
        if w > threshold:
            filtered_compounds.append(c)
            filtered_weights.append(w)

    if filtered_compounds:
        data[food_id]['compounds'] = filtered_compounds
        data[food_id]['weights'] = filtered_weights
    else:
        data[food_id]['compounds'] = [1027]
        data[food_id]['weights'] = [0.1]

#  
for food_id, info in data.items():
    print(f"{food_id}: {info}")
    
```
