## Graph 종류 
- simple Undirected Graph - G(V, E) : neighborhood(ego) graph - 해당 노드에 연결된 부분그래프 
- Directed Graph - G(V, E) : indegree, outdegree 존재
- Multi Graph - G(V, E) : 동일한 노드쌍에 대하여 여러 간선이 존재
- Weighted Graph - G(V, E, w) : edge or node에 실수 가중값 존재
- Multipartite Graph : k개의 노드집합으로 분할, 각 집합간 edge만 존재
- Adjacency Matrix

```python
import networkx as nx
# Dataset : social network of 34 members
G = nx.karate_club_graph()
type(G)
```

```python
# Visualize
# with_labels : present each node's number (default is False)
nx.draw(G, with_labels = True)
```

### Question 1: What is the average degree of the karate club network?

```python
def average_degree(num_edges, num_nodes):
    avg_degree = 0 
    avg_degree = round(2 * num_edges / num_nodes)
    return avg_degree

num_edges = G.number_of_edges()
num_nodes = G.number_of_nodes()
print(f"edge : {num_edges}, node : {num_nodes}")
avg_degree = average_degree(num_edges, num_nodes)
print(f"degree : {avg_degree}")
```
